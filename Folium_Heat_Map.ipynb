```python
import pandas as pd
import folium
Taxi_dataset=pd.read_csv("C:/users/naikp/Database/taxi_dataset.csv")
#Taxi
```

```python
Taxi_dataset.shape
```

```python
Taxi=Taxi_dataset.dropna(0)
Taxi.shape
#Taxi
```

```python
Taxi.pickup_datetime = pd.to_datetime(Taxi.pickup_datetime, format='%d-%m-%Y %H:%M')
def getMonth(x):
    return x.month
def getDay(x):
    return x.day
def getHour(x):
    return x.hour
Taxi['month'] = Taxi.pickup_datetime.apply(getMonth)
#Taxi['week'] = Taxi.pickup_datetime.apply(lambda x: x.week)
Taxi['day'] = Taxi.pickup_datetime.apply(getDay)
Taxi['hour'] = Taxi.pickup_datetime.apply(getHour)
Taxi
```

```python
from folium.plugins import HeatMap
HeatMap(data=Taxi[['pickup_latitude', 'pickup_longitude']]
        .groupby(['pickup_latitude', 'pickup_longitude']).sum().reset_index().values.tolist(), 
        radius=8,max_zoom=10).add_to(basemap)
basemap
```

```python
Taxi_hour_list = []
for hour in Taxi.hour.sort_values().unique():
    Taxi_hour_list.append(Taxi.loc[Taxi.hour == hour, ['pickup_latitude', 'pickup_longitude']]
                        .groupby(['pickup_latitude', 'pickup_longitude']).sum().reset_index().values.tolist())
```
