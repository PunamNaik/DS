```python
data=data.dropna()
data
```

```python
data.size
```

```python
latitude=data.Y[0]
longitude=data.X[0]
```

```python
# setting up map
m=folium.Map(location=[latitude,longitude],tiles="CartoDB dark_matter",zoom_start=12) 
m
```

```python
#plotting all incident locations
locations=data[['Y','X']]
locationlist = locations.values.tolist()
for i in range(len(locationlist)):   #to add markers 
    folium.CircleMarker(locationlist[i],radius=5,color='black',
                        fill=True,fill_color='yellow',
                        fill_opacity=0.7).add_to(m)
m
```

```python
print(f'total crime incidents recorded are=',len(locationlist))
```
