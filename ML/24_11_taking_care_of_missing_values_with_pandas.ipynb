```python
import pandas as pd
import numpy as np
df=pd.read_csv('C://users/naikp/Database/property data.csv')
df
```

```python
df = df.replace(to_replace=('na', '--', 'NaN'), value = np.nan)
df['OWN_OCCUPIED'] = df['OWN_OCCUPIED'].replace('12',np.nan)
df['NUM_BATH'] = df['NUM_BATH'].replace('HURLEY',np.nan)
df
```

```python
df.info()
```

```python
#checking if a column has any nulls
print(df['ST_NUM'].isnull())
```

```python
#checking if df has any null values
df.isnull().sum()
```

```python
#catching all null values in df
df.isna()
```

```python
#replacing nulls with 0 value
df.fillna(0)
```

```python
df.median()    # works on numerical data only
```

```python
df.mean()
```

```python
#replacing all missing with median of the respective column
df.fillna(df.median())   #applicable on numerical values
```

```python
#replacing all missing with mean of the respective column
df.fillna(df.mean())     # works on float
```

```python
#fill null with the value that immediately preceeded it in the dataframe
df.fillna(method='ffill')
```

```python
#To fill take the value that comes after the null
df.fillna(method='bfill')
```

```python
#Dropping an entire row, if it includes a null
df.dropna(axis=0)
```

```python
#Drop an entire column that includes a null
df.dropna(axis=1)  
```

```python
#Drop a row that has a certain percentage of nulls
df.dropna(thresh=6)  #kept only rows having 6 non-null values 
```

```python
#dropping column having nulls %
df.dropna(axis=1, thresh=8) #kept column having 8 non-na values
```
