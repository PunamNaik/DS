Apriori - data mining (Market Basket)
to find correlation, association

```python
df = pd.read_csv('C://Users/naikp/Downloads/Market_Basket_Optimisation.csv', header=None)
df.head(5)
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(df.values[i,j]) for j in range(0,20)])
transactions
```

```python
#Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
#Visualising the results
results = list(rules)
```

```python
results
```

```python
results = pd.DataFrame(list(results))
results
```

```python
list_item_base = []
list_item_add = []
list_confidence = []
list_lift = []

mylist = results['ordered_statistics'].tolist()
for i in mylist:
    list_item_base.append(i[0][0])
    list_item_add.append(i[0][1])
    list_confidence.append(i[0][2])
    list_lift.append(i[0][3])
```

```python
results['item_base'] = list_item_base
results['item_add'] = list_item_add
results['confidence'] = list_confidence
results['lift'] = list_lift
results
```
