Random variables: 2types

*continous- something that can be measured 

*discrete- something that can be counted

*distribution: uniform, binomial, gaussian,normal, gamma,exponential, poisson, etc.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
from scipy.stats import uniform
```

```python
sns.set(color_codes=True)   #setting plotting style
sns.set(rc={'figure.figsize':(5,5)})  #rc for dict format and plot size
```

# For  Continuous data following distributions can be used

# Uniform Distribution

```python
from scipy.stats import uniform
#random nos. from uniform distribution
n=10000
start=10
width=20
data_uniform= uniform.rvs(size=n, loc=start, scale=width)
```

```python
ax=sns.distplot(data_uniform, bins=100, kde=True, color='skyblue', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='uniform distribution', ylabel='Frequency')
```

# Normal Distribution

```python
from scipy.stats import norm
data_norm= norm.rvs(size=10000, loc=0, scale=1)
ax=sns.distplot(data_norm, bins=100, kde=True, color='red', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='normal/gaussian distribution', ylabel='Frequency')
```

# Gamma Distribution

```python
from scipy.stats import gamma
data_gamma= gamma.rvs(a=5, size=10000)
ax=sns.distplot(data_gamma, bins=100, kde=True, color='green', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='gamma distribution', ylabel='Frequency')
```

# Exponential Distribution

```python
from scipy.stats import expon
data_expon= norm.rvs(size=10000, loc=0, scale=1)
ax=sns.distplot(data_expon, bins=100, kde=True, color='purple', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='exponential distribution', ylabel='Frequency')
```

# For  Discrete data following distributions can be used

# Poisson Distribution

```python
from scipy.stats import poisson   
data_poisson= poisson.rvs(size=10000, mu=3)
ax=sns.distplot(data_poisson, bins=30, kde=False, color='blue', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='poisson distribution', ylabel='Frequency')
```

# Binomial Distribution

```python
from scipy.stats import binom   
data_binom= binom.rvs(size=10000, n=10, p=0.8)
ax=sns.distplot(data_binom, kde=False, color='blue', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='binomial distribution', ylabel='Frequency')
```

# Bernoulli Distribution

```python
from scipy.stats import bernoulli  
data_bern= bernoulli.rvs(size=10000, p=0.6)
ax=sns.distplot(data_bern, kde=False, color='blue', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='bernoully distribution', ylabel='Frequency')
```
