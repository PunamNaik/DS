# Fetching a random image from web and displaying a message with frame on top of it

```python
url='https://picsum.photos/500/500?random'
response=requests.get(url)
img=response.content
img=io.BytesIO(img)
img=Image.open(img).convert('RGB')
from PIL import ImageOps
img=ImageOps.expand(img, border=5, fill='blue')
img
```

```python
newImg=Image.new('RGB', ((img.size[0]+20),(img.size[1]+50)), 'Black')
#background                     
```

```python
width,height=img.size
newImg.paste(img, (10,20))
newImg
```

```python
from PIL import ImageDraw, ImageFont
img_draw=ImageDraw.Draw(newImg)
center=(newImg.size[0]/2)-110
font=ImageFont.truetype('E:/Software/matlab/sys/java/jre/win64/jre/lib/fonts/LucidaBrightDemiBold.ttf', 40)
img_draw.text((center,100), "Hello Pillow", fill='Red', font=font)
newImg
```
