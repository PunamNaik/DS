```python
#checking if column has any nulls
print(df['ST_NUM'].isnull())
```

```python
#checking if df has any null values
df.isnull().sum()
```

```python
#catching all null values in df
df.isna()
```

```python
#replacing nulls with 0 value
df.fillna(0)
```

```python
#replacing all missing with median of the respective column
df.fillna(df.median())   #applicable on numerical values data
```

```python
#replacing all missing with median of the respective column
df.fillna(df.mean())    #applicable on numerical values data
```

```python
#df.groupby(df.columns, axis = 1).transform(lambda x: x.fillna(x.mean()))
```

```python
#fill null with the value that immediately preceeded it in the dataframe
df.fillna(method='ffill')
```

```python
#To fill take the value that comes after the null
df.fillna(method='bfill')
```

```python
#Dropping an entire row, if it includes a null
df.dropna(axis=0)
```

```python
#Drop an entire column that includes a null
df.dropna(axis=1)
```

```python
#Drop a row that has a certain percentage of nulls
df.dropna(thresh=6)  #kept only rows having 6 non-null values 
```

```python
#dropping column having nulls %
df.dropna(axis=1, thresh=8) #kept column having 8 non-na values
```
