```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#top 5 wine producing countries
plt.figure(figsize=(15,10))
country.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("country of origin")
plt.ylabel("no. of wines")
plt.show()
```

```python
#top quality wine producing countries
plt.figure(figsize=(20,15))
country.max().sort_values(by='points',ascending=False)['points'].plot.bar()
plt.xticks(rotation=50)
plt.xlabel("country of origin")
plt.ylabel("no. of points")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
#create and generate a word cloud image
text=df.description[0]
wordcloud=WordCloud().generate(text)
```

```python
#display
plt.imshow(wordcloud, interpolation='bilinear')  #interpolation for image appears more smoothly
plt.axis('off')
plt.show()
```

```python
#changing visuals
wordcloud=WordCloud(max_font_size=50, max_words=100, background_color='white').generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')  #interpolation for image appears more smoothly
plt.axis('off')
plt.show()
```

```python
#to save the image in folder
wordcloud.to_file('WordCloudImage#1.png')
```

```python
#skipping words which not required
stopwords=set(STOPWORDS)
stopwords.update(["drink","now","wine","include","dried"])
wordcloud=WordCloud(stopwords=stopwords, max_font_size=50, max_words=100, background_color='white').generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')  #interpolation for image appears more smoothly
plt.axis('off')
plt.show()
```
