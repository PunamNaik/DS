```python
import seaborn as sns
sns.set()
flights=sns.load_dataset("flights")
flights
```

```python
flights=flights.pivot("month","year","passengers")
ax=sns.heatmap(flights)
```

```python
ax=sns.heatmap(flights,annot=True,fmt='d')
#annot-annotate each cell with numeric value using integer formatting
```

```python
ax=sns.heatmap(flights, linewidths=.5)
```

```python
ax=sns.heatmap(flights, cmap="YlGnBu")  # cmap-for diff color
```

```python
ax=sns.heatmap(flights, center=flights.loc["January",1960])
```

```python
ax=sns.heatmap(flights, cbar=False)  #no color bar on side
```

```python
import numpy as np
data=np.random.randn(50,20)#50*20 metrix
#plot every other & dont plot row labels if used yticklabels=False
ax=sns.heatmap(data, xticklabels=2, yticklabels=2)  
```

```python
#use a mask to plot only part of matrix
corr=np.corrcoef(np.random.randn(10,200))
mask=np.zeros_like(corr)
mask[np.triu_indices_from(mask)]=True
with sns.axes_style("white"):
    ax=sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```
